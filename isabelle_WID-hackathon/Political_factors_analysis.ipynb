```python
import pandas as pd
import matplotlib, matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('aggregate_political_dataset.csv', sep='\t')
```

```python
# Scatter plots of political factors v. % of cases and deaths relative to population

data.columns
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,4))

ax[0].scatter(data.Poli_stability, data.Case_percent)
ax[0].set(ylabel='Cases (% of population)', xlabel='Political stability',ylim=(0,5))
ax[1].scatter(data.Poli_stability, data.Death_percent)
ax[1].set(ylabel='Deaths (% of population)', xlabel='Political stability',ylim=(0,0.15))
plt.tight_layout()
```
