```python
import pandas as pd
import seaborn as sns
import matplotlib, matplotlib.pyplot as plt
```

```python
covdata = pd.read_csv('COVID19 Hackathon Dataset.csv')

# to teamnatalia - i have uploaded the dataset as well
lockdowndata = pd.read_excel('../../Downloads/acaps_covid19_government_measures_dataset_0.xlsx', sheet_name='Dataset')
```

```python
lock_date = lockdowndata.loc[:,['COUNTRY', 'CATEGORY', 'DATE_IMPLEMENTED']]
```

```python
lock_date.rename(columns={'COUNTRY':'Countries'}, inplace=True)
```

```python
covdf = covdata.fillna(0)
```

```python
covdf['Date'] = pd.to_datetime(covdf['Date'], format="%Y/%m/%d")
```

```python
lock_date.columns
```

```python
merged = pd.merge(lock_date, covdf, on='Countries')
```

```python
lockdown_category = merged[merged.CATEGORY == 'Lockdown']
```

```python
lockdown_start_df = lockdown_category.groupby('Countries').agg(lockdown_starts = ('DATE_IMPLEMENTED', min))
```

```python
lockdown_start_df.reset_index(inplace=True)
```
